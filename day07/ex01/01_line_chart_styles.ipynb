```python
import pandas as pd
import sqlite3

db_con = sqlite3.connect('../checking-logs.sqlite.sqlite')
```

• analyze only the users and not the admins<br>
• analyze only the dates when there were both views and checker commits<br>
• use size of the font should be 8<br>
• the size of the figure is (15,8)

```python
commits = pd.io.sql.read_sql("select timestamp from checker WHERE uid LIKE 'user_%'", db_con)
commits = pd.DataFrame(pd.to_datetime(commits['timestamp']).dt.date)
commits.rename(columns={'timestamp':'commits'}, inplace=True)
commits = pd.DataFrame(commits.groupby('commits')['commits'].count())
commits.index.name = 'date'
commits
```

```python
views = pd.io.sql.read_sql("select datetime from pageviews WHERE uid LIKE 'user_%'", db_con)
views = pd.DataFrame(pd.to_datetime(views['datetime']).dt.date)
views.rename(columns={'datetime':'views'}, inplace=True)
views = pd.DataFrame(views.groupby('views')['views'].count())
views.index.name = 'date'
views
```

```python
graph = pd.merge(views, commits, on='date', how='inner')
graph
```

```python
# color=['r', 'xkcd:aqua']
graph.plot(fontsize=8, figsize=(15,8), title='Views and commits per day', rot=90, style=['o-.', '--'], color=['r', 'xkcd:aqua']).grid(axis='x')
```
